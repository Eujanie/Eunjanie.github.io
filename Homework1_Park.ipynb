####<font color=red>Task 2</font>  (3 points)

1. Use the data munging techniques from lecture to create a dataframe called `states`
  * The dataframe should have a row for every state:
    * Exclude Puerto Rico
    * You may include Washington DC
  * The dataframe should have 8 columns
    * state
    * population change
    * natural increase
    * births
    * deaths
    * migration total
    * international migration
    * domestic migration
2. Show the `head()` and `tail()` of the dataframe `states` to confirm the data is formatted properly


```
states <- events[-(1:8),]
head(states, 10)
```

<br>

---

<br>






```
nb_row <- nrow(states)
(nb_row-11):nb_row
```

```
states <- states[-((nb_row-11):nb_row),]
tail(states)
```

```
nrow(states)
summary(states[, 9:12])
```

```
states <- states[, -(9:12)]
head(states)
```

```
names(states) <- c("state", "totalpopchange", "naturalincrease", "births", "deaths", "migrationtotal","internationalmigration", "domesticmigration" )
head(states)
```

```
states$state <- gsub('\\.', '', states$state)
head(states)
```

```
states$totalpopchange <- as.numeric(states$totalpopchange)
states$naturalincrease <- as.numeric(states$naturalincrease)
states$births <- as.numeric(states$births)
states$deaths <- as.numeric(states$deaths)
states$migrationtotal <- as.numeric(states$migrationtotal)
states$internationalmigration <- as.numeric(states$internationalmigration)
states$domesticmigration <- as.numeric(states$domesticmigration)

head(states)
```

```
rownames(states) <- NULL
head(states)
tail(states)
```
